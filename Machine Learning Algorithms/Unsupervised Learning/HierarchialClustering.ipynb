```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=800,n_features=4,centers=6,
                  cluster_std=1.5,random_state=42)

sns.scatterplot(x=X[:,0], y=X[:,1], c=y);
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(10,7))
plt.title('Dendrogram')
plt.xlabel('')
sch.dendrogram(sch.linkage(X, method='ward'));
```

> **How to find optimum number of clusters?**
>> To find optimum number of clusters we count the number of vertical lines our horizontal line is intersecting. This horizontal line is chosen such that it intersects  the longest vertical line without any other horizontal line intersecting it.

```python
sns.clustermap(X,figsize=(10,6),method='ward');
```

```python
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(n_clusters=6,metric='euclidean')

agc.fit_predict(X);
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))

sns.scatterplot(x=X[:,0], y=X[:,1], c=y, ax=ax1)
ax1.set_title('Actual')

sns.scatterplot(x=X[:,0], y=X[:,1], c=agc.labels_, ax=ax2)
ax2.set_title('Predicted');
```
