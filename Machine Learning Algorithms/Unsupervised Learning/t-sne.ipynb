# t - SNE : t - distributed Stochastic Neighbor Embedding

t-SNE is a dimensionality reduction technique commonly used for visualizing high-dimensional data in a lower-dimensional space, typically 2D or 3D. It works by modeling each high-dimensional data point with a probability distribution in the lower-dimensional space, aiming to preserve the local structure of the data. t-SNE is particularly effective at revealing the underlying structure and relationships within complex datasets, making it a popular tool for exploratory data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
link = 'https://raw.githubusercontent.com/daaanishhh002/MachineLearning/main/Datasets/digits.csv'
df = pd.read_csv(link).dropna()

df.shape
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
tsne = TSNE(perplexity=20)
pca = PCA(n_components=2)

X_tsne  = tsne.fit_transform(X)
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape, X_tsne.shape
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(15,6))

sns.scatterplot(x=X_tsne[:,0],y=X_tsne[:,1],c=y,cmap='tab10',ax=ax1)
ax1.set_title('t-sne Implementation')

sns.scatterplot(x=X_pca[:,0],y=X_pca[:,1],c=y,cmap='tab10',ax=ax2)
ax2.set_title('PCA Implementation');
```

```python
import plotly.express as px
px.scatter(x=X_tsne[:,0],y=X_tsne[:,1],color=y.astype(str),title='t-sne Implementation')
```
