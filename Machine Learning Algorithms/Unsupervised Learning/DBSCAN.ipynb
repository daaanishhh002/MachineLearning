# DBSCAN : Density Based Spatial Clustering of Applications with Noise

DBSCAN is a clustering algorithm that groups data points based on their density. It doesn't require specifying the number of clusters beforehand and is good at handling clusters of different shapes, while also being robust to noise. It categorizes points as core, border, or noise based on their density distribution and expands clusters from core points.

![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/db6-e1584577503359.png)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=500,noise=0.06,factor=0.4)

sns.scatterplot(x=X[:,0],y=X[:,1],c=y);
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.1,min_samples=10)

db.fit_predict(X);
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))

sns.scatterplot(x=X[:,0], y=X[:,1], c=y, ax=ax1)
ax1.set_title('Actual')

sns.scatterplot(x=X[:,0], y=X[:,1], c=db.labels_, ax=ax2)
ax2.set_title('Predicted');
```
