# Feature Scaling

Feature scaling is a data pre-processing technique used to transform the values of features or variables in a dataset to a similar scale. The purpose is to ensure that **all features contribute equally to the model and to avoid the domination of features** with larger values.

![](https://miro.medium.com/v2/resize:fit:890/1*9j33xl7N74uvn2uy1xqG7A.png)

## Feature Scaling Techniques

1. Normalisation
2. Standardisation

## Normalisation

Normalization is the process of scaling numeric features to a standard range, like 0 to 1. It helps prevent features with large values from dominating those with smaller values during model training, ensuring fair comparison and better performance.

### Normalisation Techniques

1. MinMax Scaling
2. Mean Normalisation
3. Max Absolute
4. Robust Scaling

 #### MinMax Scaling
 
 Most famous and widely used technique. Used when we are sure about the minimum and maximum of a certain parameter.

$$\hat{x_i} = \frac{x_i - x_{min}}{x_{max} - x_{min}}$$

```python
link = 'https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv'
df = pd.read_csv(link,usecols=[0,1,2])
df.columns = ['Class Label','Alcohol','Malic Acid']

df.sample(5)
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:,1:]
y = df.iloc[:,0]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2002)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train) # learning the parameters

# transforming the sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.data_max_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.scatter(X_train['Alcohol'],X_train['Malic Acid'],c=y_train)
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Alcohol'],X_train_scaled['Malic Acid'],c=y_train)
ax2.set_title('After Scaling')

fig.supxlabel('Alcohol')
fig.supylabel('Malic Acid');
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'],ax=ax1)
sns.kdeplot(X_train['Malic Acid'],ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(X_train_scaled['Malic Acid'],ax=ax2)

plt.show()
```

#### Mean Normalisation

$$\hat{x_i} = \frac{x_i - x_{mean}}{x_{max} - x_{min}}$$

#### Max Abs Scaling

Used when sparse data is present.

$$\hat{x_i} = \frac{x_i}{|x_{max}|}$$

#### Robust Scaling

Robust to outliers.

$$\hat{x_i} = \frac{x_i - x_{median}}{IQR}$$

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
X_train_scaled.describe()
```

```python
X_train.describe()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.scatter(X_train['Age'],X_train['EstimatedSalary'],c=y_train)
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],c=y_train)
ax2.set_title('After Scaling');
```

## Standardisation (z-score normalisation)

We are centring the data around the mean($\mu$) and scaling it by the standard deviation($\sigma$). This distributes the data normally and the **transformed data has normal characteristics**.

$$z_i = \frac{x_i-\bar{x}}{\sigma}$$

- Shape doesn't change.
- Outliers don't get affected.

```python
link = 'https://raw.githubusercontent.com/shivang98/Social-Network-ads-Boost/master/Social_Network_Ads.csv'
df = pd.read_csv(link)

df = df.iloc[:,-3:]
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased',axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3, random_state=2002)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.var_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.scatter(X_train['Age'],X_train['EstimatedSalary'],c=y_train)
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],c=y_train)
ax2.set_title('After Scaling');
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train,ax=ax1)
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled,ax=ax2);
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'],ax=ax1)
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'],ax=ax2);
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()

dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
print(f'Actual {accuracy_score(y_test,y_pred)}')
print(f'Scaled {accuracy_score(y_test,y_pred_scaled)}')
```

```python
dt.fit(X_train,y_train)
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print(f'Actual {accuracy_score(y_test,y_pred)}')
print(f'Scaled {accuracy_score(y_test,y_pred_scaled)}')
```

### Where to use Standardisation?

1. K-Means
2. K-Nearest Neighbours
3. Principal Component Analysis
4. Artificial Neural Networks
5. Gradient Descent

## Normalisation vs Standardisation

- Use standardization when the data has outliers and follows a Gaussian distribution.
- Use normalization when the distribution is not Gaussian and when the scale of the features matters, but be cautious of outliers.
